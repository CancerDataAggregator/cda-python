## Data frame conversion
Added functionality that allows users to convert the json data into pandas data frame. Also allows subsetting of nested fields.

```python
q = Q('id = "TCGA-E2-A10A"')
df = q.run().to_dataframe()
df.head()
```

```python
q = Q('id = "TCGA-E2-A10A"')
r = q.run()
df = r.to_dataframe(
    record_path=['identifier'],
    meta=['id', 'species', 'sex'])
df.head()
```

## Disable Q() text output
Enables users to hide the output of a query run for use cases that involve looping over many queries.

```python
q1 = Q('ResearchSubject.Diagnosis.stage = "Stage IIIC"')
r = q1.run(verbose=False)
r
```

## Filtering
Allows users to reduce the amount of data returned to a desired set of fields

```python
q = Q('ResearchSubject.Diagnosis.stage = "Stage IIIC"')
r = q.run(filter='id, ResearchSubject.identifier, ResearchSubject.identifier.system')
r.pretty_print()
```

## FROM, OR operators

```python
disease1 = Q('ResearchSubject.primary_diagnosis_site = "Ovary"')
disease1.run()
```

```python
disease2 = Q('ResearchSubject.primary_diagnosis_site = "Breast"')
disease2.run()
```

```python
disease1 = Q('ResearchSubject.primary_diagnosis_site = "Ovary"')
disease2 = Q('ResearchSubject.primary_diagnosis_site = "Breast"')
q = disease1.Or(disease2)
r = q.run()
r
```

```python
q1 = Q('ResearchSubject.primary_diagnosis_condition = "Ovarian Serous Cystadenocarcinoma"')
q2 = Q('ResearchSubject.identifier.system = "PDC"')
r = q1.And(q2).run()
r
```

```python
q1 = Q('ResearchSubject.primary_diagnosis_condition = "Ovarian Serous Cystadenocarcinoma"')
q2 = Q('ResearchSubject.identifier.system = "PDC"')
q3 = Q('ResearchSubject.identifier.system = "GDC"')

q = q3.From(q1.And(q2))
r = q.run()
r
```
